```python
bseat = set()
mseat = set()
class Theater:
    def __init__(self,bseat,mseat):
        self.bseat=bseat
        self.mseat=mseat
    def seat_availability(self):
        print(f"Booked seats in middle class are {self.mseat}\nBooked seats in balcony are {self.bseat}")
        

class Details(Theater):
    def __init__(self):
        self.name = input("Enter your name: ")
        self.city = input("\nEnter your city name: ")
        self.tname = input("Enter the theater name: ")
        self.tcity = input("Enter the name of city: ")
class Payment:
    def __init__(self, pay, qty, ticket_price, new, seat_type):
        self.pay = pay
        self.qty = qty
        self.ticket_price = ticket_price
        self.new = new
        self.balance = 10000  
        self.seat_type = seat_type
        self.status = ""
        self.book_seats()

    def book_seats(self):
        if self.balance >= (self.qty * self.ticket_price):
            self.balance -= self.qty * self.ticket_price
            if self.seat_type == "Balcony":
                bseat.update(self.new)
                globals()[self.tname]=Theater(bseat,0)
            else:
                mseat.update(self.new)
                globals()[self.tname]=Theater(0,mseat)
            print("Booking successful")
            self.status = "Booked successfully"
        else:
            self.status = "Unsuccessful"
            print("You don't have enough balance")

class TicketBooking(Details, Payment):
    def __init__(self):
        Details.__init__(self)
        self.type = input("Enter the seat type (Balcony/Middle class): ").capitalize()
        if self.type == "Balcony":
            ticket_price = 100
        else:
            ticket_price = 80

        print(f"{self.type} ticket price: {ticket_price}")
        print(f"Occupied seats in {self.type} section:", end="\n")
        for s in (bseat if self.type == "Balcony" else mseat):
            print(s, end=", ")
        
        new = input("Enter the seat number(s) separated by commas: ").split(",")
        new = [int(s.strip()) for s in new]
        new.sort()
        empty = 0
        for i in new:
            if i in (bseat if self.type == "Balcony" else mseat):
                empty += 1
        
        if empty != 0:
            print("Some of the seats are already occupied. Try with another seat number.")
        else:
            qty = len(new)
            pay = input("Enter the UPI ID (e.g., Gpay, Phonepay, Paytm): ")
            print("Fetching UPI details...")
            Payment.__init__(self, pay, qty, ticket_price, new, self.type)

    def booking_details(self):
        print(f"NAME: {self.name}\nADDRESS: {self.city}\nSEAT TYPE: {self.type}\nQTY: {self.qty}\nTRANSACTION DETAILS: {self.pay}\nPAYMENT STATUS: {self.status}\nAccount Balance: {self.balance}")
        print("Thank you!\n")

    def seat_availability(self):
        print(f"Available Balcony seats: {[i for i in range(1, 21) if i not in bseat]}")
        print(f"Available Middle class seats: {[i for i in range(1, 21) if i not in mseat]}")

import os

choice = 0
while choice != 5:
    print("MENU\n1 --> Book Tickets\n2 --> Cancel booking\n3 --> Check seat availability\n4 --> Booking Details\n5 --> EXIT")
    choice = int(input("Enter your choice: "))
    if choice == 1:
        def create_var(var_name):
            globals()[var_name] = TicketBooking()
        name = input('Username:\n')
        create_var(name)
    elif choice == 2:
        pass
    elif choice == 3:
        try:
            name = input("Enter username: ")
            globals()[name].seat_availability()
        except KeyError:
            print(f"Booked seats in middle class are 0\nBooked seats in balcony are 0")

    elif choice == 4:
        name = input("Enter username: ")
        try:
            globals()[name].booking_details()
        except:
                print("This username don't have any booking details")
    else:
        os._exit(0)
os._exit(0)
```
